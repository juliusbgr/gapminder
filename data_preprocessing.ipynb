```python
import pandas as pd

df_pop = pd.read_csv("app/pop.csv")
df_life = pd.read_csv("app/lex.csv")
df_gni = pd.read_csv("app/gnipercapita_ppp_current_international.csv")
```

```python
def to_tidy(df, value_name):
    tidy = (
        df.melt(id_vars=["country"],var_name="year", value_name=value_name)
        
    )
    tidy["year"] = tidy["year"].astype(int)
    return tidy

clean_pop  = to_tidy(df_pop,  "population")
clean_life = to_tidy(df_life, "lifeExp")
clean_gni  = to_tidy(df_gni,  "gniPPP")
```

```python
clean_pop.head(5)
```

```python
def parse_number(val):
    """
    Convert strings to floats.
    Return the value unchanged if it's None, NaN, non-string, or parsing fails.
    """
    if pd.isna(val) or not isinstance(val, str):
        return val

    s = val.strip().replace(",", "")
    multipliers = {"k": 1e3, "M": 1e6, "B": 1e9}

    try:
        suffix = s[-1]
        if suffix in multipliers:
            return float(s[:-1]) * multipliers[suffix]
        return float(s)
    except (ValueError, TypeError):
        return val
```

```python
clean_pop['population'] = clean_pop['population'].apply(parse_number)
clean_gni['gniPPP'] = clean_gni['gniPPP'].apply(parse_number)
clean_life['lifeExp'] = clean_life['lifeExp'].apply(parse_number)
```

```python
clean_pop.head(10)
```

```python
# clean population
clean_pop = clean_pop.sort_values(['country', 'year'])
clean_pop['population'] = (
    clean_pop.groupby('country')['population']
             .apply(lambda s: s.ffill().bfill())
             .reset_index(level=0, drop=True)
)

# clean life expectancy
clean_life = clean_life.sort_values(['country', 'year'])
clean_life['lifeExp'] = (
    clean_life.groupby('country')['lifeExp']
              .apply(lambda s: s.ffill().bfill())
              .reset_index(level=0, drop=True)
)

# clean GNI
clean_gni = clean_gni.sort_values(['country', 'year'])
clean_gni['gniPPP'] = (
    clean_gni.groupby('country')['gniPPP']
             .apply(lambda s: s.ffill().bfill())
             .reset_index(level=0, drop=True)
)
```
